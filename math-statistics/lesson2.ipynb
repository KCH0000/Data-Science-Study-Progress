```python
import numpy as np
from math import factorial as fl
from math import exp as e
```

<i>Введем фунции для урощенного рассчета
    <ul>
        <li> C - Сочетания </li>
        <li> Binom_P - Биномиальное распределение</li>
        <li> Poisson_P - Распределение Пуассона </li> 
    </ul>
</i>

```python
def C(n, k):
    """
    Функция возвращает кол-во сочетаний(без учета последовательности) к элементов из n возможных
    n - всего элементов 
    k - размер выборки из всех элементов
    """
    return int(fl(n) / (fl(k) * fl(n - k)))

def Binom_P(n, k, p):
    """
    Биномильный закон распределния вероятносей
    Функция возвращет вероятность k наступлени события с А в n независимых испытаниях
    n - количество испытаний
    k - количество наступления события A из n испытаний
    p - вероятность настуления А в каждом из испытаний
    """
    q = 1 - p
    m = n - k
    return C(n, k) * p**k * q**m

def Poisson_P(n, m, p):
    """
    Распределение Пуассона
    Функция возвращет вероятность k наступлени события с А в n независимых испытаниях
    когда веростность наступления А мала, а количество испытаний большое
    n - количество испытаний
    m - количество наступления события A из n испытаний
    p - вероятность настуления А в каждом из испытаний
    """
    lambda_ = n * p
    return (lambda_**m / fl(m)) * e(-lambda_)
```

<h2>Задание 1 </h2> <br>
<i>Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.<i>

<b>Решение:</b> Вероятность попадания высока, поэтому применим формулу биномиального распределения, где<br> 
p = 0.8<br>
n = 100<br>
k = 85

```python
P = Binom_P(100, 85, 0.8)
print(f'Ответ: Вероятность = {np.round(P, 3)}')
```

<h2>Задание 4 </h2> <br>
<i>
В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча.<br>
4.1 Какова вероятность того, что все мячи белые?<br>
4.2 Какова вероятность того, что ровно два мяча белые?<br>
4.3 Какова вероятность того, что хотя бы один мяч белый?
<i>

<b>Решение 4.1:</b>
Чтобы все шары были белые, нужно вытащить 2 белых шара из корзины 1 и 2 белых шара из корзины 2. Нужно посчитать вероятность для первой и второй корзины. Т.к. нам нужно, что бы эти два события произошли одновременно, то нужно взять произведение вероятностей.

```python
P_1 = C(7, 2) / C(10, 2)
P_2 = C(9, 2) / C(11, 2)
P = P_1 * P_2
print(f'Ответ: Вероятность = {np.round(P, 3)}')
```

<b>Решение 4.2:</b>
Рассмотрим случаи когда,<br>
вытащили 0 белых из первой корзины, 2 из второй<br>
вытащили 2 белых из первой корзины, 0 из второй<br>
вытащили 1 белых из первой корзины, 1 из второй<br>

```python
P = 0
for i in range(3):
     p_1_bucket = C(7, i) * C(3, 2 - i) / C(10, 2)
     p_2_bucket = C(9, 2 - i) * C(2, i) / C(11, 2)
     P +=  p_1_bucket * p_2_bucket
print(f'Ответ: Вероятность = {np.round(P, 3)}')
```

<b>Решение 4.3:</b>
В данном случае проще посчитать, когда все шары НЕ белые(цвет остальных мы не знаем :)). Это событие несовместно с событием, когда хотя бы 1 белый и полная веротяность обоих событий равна 1.<br>
В корзине 1 !белых 3 шт.<br>
В корзине 2 !белых 2 шт.<br>

```python
P_1 = C(3, 2) / C(10, 2)
P_2 = C(2, 2) / C(11, 2)
P = 1 - P_1 * P_2
print(f'Ответ: Вероятность = {np.round(P, 3)}')
```
