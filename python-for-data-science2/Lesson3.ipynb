<h2>Задание 1</h2>
Для чего и в каких случаях полезны различные варианты усреднения лейблов целевой переменной: micro, macro, weighted?

***Ответ:***<br>
Для начала разберемся в математическом смысле в различных стратегиях мульклассовой F-меры.<br>
***Macro*** стратегия это невзвещенное устредненное значение показателя. Для каждого класса вычисляется полнота или точно, а затем берется среднее значение для всех классов.<br>
***Micro*** стратегия это когда вычисялется общие TP, TN, FP, FN по всем классам, а уже после вычисляется полнота и точность.<br>
***Weighted*** стратения это взвешенное устредненное значение показателя. Уровень поддержки каждого класса рассчитавается из количества примеров класса.<br>
<br>
    Если стоимость ошибки по каждому классу равнозначна то нужно использовать Macro стратегию, т.к. она дает однинаковый вес для каждого класса. Иначе нужно использовать micro стратегию, где одинаковый вес у каждого примера. Чтобы увеличить вес класса при micro стратегии, то в необходимо увеличить в тестовой выборке количество примеров данного класса.
