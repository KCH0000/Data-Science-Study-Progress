```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 12, 9
plt.style.use('fivethirtyeight')
```

```python
cards = pd.read_csv('lesson6/creditcard.csv')
cards.reset_index(drop=True)
```

```python
class_count = cards['Class'].value_counts()
```

```python
class_count.plot('bar')
plt.show()
```

```python
class_count.plot(kind='bar', logy=True)
plt.show()
```

```python
v1_class_0 = cards.loc[cards['Class'] == 0, ['V1']]
v1_class_1 = cards.loc[cards['Class'] == 1, ['V1']]
```

```python
names = ['Class 0', 'Class 1']
colors = ['gray', 'red'] 
```

```python
plt.hist([v1_class_0['V1'], v1_class_1['V1']], bins=20,
         histtype='stepfilled', density=True, alpha=0.5,
         color=colors, label=names)
plt.legend(loc='center left')
plt.xlabel('Class')
plt.show()
```
