```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
k = 3
plt.plot(x, np.cos(x))
plt.plot(x, np.cos(x * k))
```

```python
R = 5
x = np.linspace(-R, R, 1000)
plt.plot(x, np.sqrt(R**2 - x**2))
plt.plot(x, -np.sqrt(R**2 - x**2))
```

```python
A = 5
B = 10
x = np.linspace(-A, A, 1000)
plt.plot(x, np.sqrt((B**2 * (A**2 - x**2))/A**2))
plt.plot(x, -np.sqrt((B**2 * (A**2 - x**2))/A**2))
```

```python
A = 5
B = 10
x1 = np.linspace(-A * 10, -A, 1000)
x2 = np.linspace(A, A * 10, 1000)
plt.plot(x1, np.sqrt((B**2 * (-A**2 + x1**2))/A**2))
plt.plot(x2, -np.sqrt((B**2 * (-A**2 + x2**2))/A**2))
plt.plot(x1, -np.sqrt((B**2 * (-A**2 + x1**2))/A**2))
plt.plot(x2, np.sqrt((B**2 * (-A**2 + x2**2))/A**2))
```
