```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
k = 3
plt.plot(x, np.cos(x))
plt.plot(x, np.cos(x * k))
```

```python
R = 5
x = np.linspace(-R, R, 1000)
plt.gca().set_aspect('equal')
plt.plot(x, np.sqrt(R**2 - x**2))
plt.plot(x, -np.sqrt(R**2 - x**2))
```

```python
A = 5
B = 10
x = np.linspace(-A, A, 1000)
plt.gca().set_aspect('equal')
plt.plot(x, np.sqrt((B**2 * (A**2 - x**2))/A**2))
plt.plot(x, -np.sqrt((B**2 * (A**2 - x**2))/A**2))
```

```python
A = 5
B = 10
x1 = np.linspace(-A * 10, -A, 1000)
x2 = np.linspace(A, A * 10, 1000)
plt.plot(x1, np.sqrt((B**2 * (-A**2 + x1**2))/A**2))
plt.plot(x2, -np.sqrt((B**2 * (-A**2 + x2**2))/A**2))
plt.plot(x1, -np.sqrt((B**2 * (-A**2 + x1**2))/A**2))
plt.plot(x2, np.sqrt((B**2 * (-A**2 + x2**2))/A**2))
```

```python
def f(k, a, b, x):
    return k * np.cos(x - a) + b
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
for k in range(-10, 11, 10):
    for a in range(-10, 11, 10):
        for b in range(-10, 11, 10):
            plt.plot(x, f(k, a, b, x))
show()
```

```python
x = np.linspace(-2, 5, 1000)
plt.plot(x, x**2 - 1, label='x2')
plt.plot(x, (np.exp(x) + x - 1) / x, label='exp')
plt.legend()
plt.show()
```

```python
from scipy.optimize import fsolve
def eq(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)
x1, y1 = fsolve(eq, (-1, 5))
x2, y2 = fsolve(eq, (2, 3.5))
x3, y3 = fsolve(eq, (3.5, 5))
print(f'Корень 1: {x1}, {y1}')
print(f'Корень 2: {x2}, {y2}')
print(f'Корень 3: {x3}, {y3}')
```
