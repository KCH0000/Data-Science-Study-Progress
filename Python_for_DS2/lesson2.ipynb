<h2>Задание 1</h2> <bn>
В чём различие между зависимыми и независимыми выборками?

***Ответ:*** <br>
Выборки называются ***зависимыми***, когда можно установить взаимооднозначную связь между случаями в выборках.<br>
Выборки называются ***независимыми***, когда связь между случаями установить невозможно.

<h2>Задание 2</h2> <bn>
Когда применяются параметрические статистические критерии, а когда их непараметрические аналоги?

***Ответ:***<br>
Параметрические критерии применяются, когда есть информация о характере распредления величины и достаточный объем выборки. Когда нет предоложений об характере распределения или выборка очень малая(грубо говоря данные низкого качества), то применяется непараметричесий аналог статистического критерия. <br>
Например:<br>
В задаче сравнения независимых выборок можно использовать t-критерий, если выборки удовлетворяют по кол-ву и нормальности распределения. Если нет, то применяется непараметрический аналог, например, критерий Манна-Уитни и двухвыборочный критерий Колмогорова-Смирнова
В задаче сравнения зависимой выборки, можно использовать t-критерий для зависимых выборок, если они опять же удолетворяют условиям применения критерия. Иначе использует непараметрические аналоги - критерий знаков и критерий Вилкоксона парных сравнений.
