<h2>Задание 1</h2> <bn>
Привести по 2 примера, когда лучше максимизировать Precision, а когда Recall?

<center>$\large precision = \LARGE \frac{TP}{TP \: + \: FP}$<center>

***"Точность"*** это доля правильных предсказаний среди всех положительных предсказаний
Приоритет на точность необходим в том случае, когда критическим является ложное срабатывание.
Например обработка фото с нарушениями должны обрабатываться в высокой точность, чтобы было как можно меньше споров о правомерности, при этом не является критическим пропуск каких-либо нарушителей. Или решение о выдачи кредита, выгодно выдавать благонадёжным людям, при не является критической ошибкой в отказе какому то проценту людей.

<center>$\large recall = \LARGE \frac{TP}{TP \: + \: FN}$<center>

***"Полнота"*** это доля правильно распознаны объектов от всех целевых объектов.<br>
Приоритет на Полноту необходим в том случае, когда пропуск целевого объекта критичен, при этом отнесение неверного объекта к целевому не является критической ошибкой, т.к. после можно провести дополнительную проверку.<br>
Первое, что приходит в голову это медицина, когда нужно, как можно меньше пропустить целевых объектов, а в случае ошибочного отнесения направить на дополнительное обследование. Второе это выявление брака на производстве, где отбраковку можно направить ручную проверку, а детали прошедшие первичный анализ гарантировано с максимальным процентом без брака. 

<h2>Задание 2</h2> <bn>
Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

<center>$\LARGE F_{\beta} = (1+ \beta^2) \cdot \frac{precision \: \cdot \: recall}{(\beta^2 \cdot precision) \: + \: recall}$<center>

F-мера это гармоническое среднее между полнотой и точность. У гармонического среднего есть свойство: если один из членов ряда стремится к нулю, то и гармоническое среднее стремится к нулю, т.е. таким образом мы можем выявить, что один и показателей имеет недостаточную величину. В случае с простым арифметическим средним один показатель может компенсироваться за счет другого.
