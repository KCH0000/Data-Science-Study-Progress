```
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
```

```
%matplotlib inline
rcParams['figure.figsize'] = 12, 9
```

```
title = 'График f(t)'
g_color = 'g'
x_label = 'Значения t'
y_label = 'Значения f'
x_range = (0.5, 9.5)
y_range = (-2.5, 2.5)
```

```
t = np.linspace(0, 10, 51)
print(t.__len__())
t
```

```
f = np.cos(t)
f
```

```
plt.axis(x_range + y_range)
plt.plot(t, f, color=g_color, label='cos(x)')
plt.title(title)
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.show()
```
