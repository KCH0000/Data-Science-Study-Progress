```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 12, 9
```

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1, y1_title = x**2, 'x**2'
y2, y2_title = 2 * x + 0.5, '2 * x + 0.5'
y3, y3_title = -3 * x - 1.5, '-3 * x - 1.5'
y4, y4_title = np.sin(x), 'sin(x)'
space = 0.3
size = (8, 6)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)

ax1.set_title(y1_title)
ax2.set_title(y2_title)
ax3.set_title(y3_title)
ax4.set_title(y4_title)

ax1.set_xlim(-5, 5)

fig.set_size_inches(size[0], size[1])
plt.subplots_adjust(wspace=space, hspace=space)
```
